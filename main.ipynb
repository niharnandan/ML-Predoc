```python
#Load modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import seaborn as sns
plt.style.use('seaborn')
```

```python
#Load cleaned dataset
data = pd.read_csv('data/data_clean.csv')
age = data['age']
#First 5 rows for reference
data.head()
```

```python
#Plot age distribution
plt.hist(age, edgecolor='white', linewidth=0.8, density=True, bins=20)
sns.kdeplot(age, linewidth=2.0)
plt.xticks(range(0,101,5))
plt.xlabel('Age')
plt.ylabel('Normalized Frequency')
plt.title('Normalized age distribution of population')
plt.savefig('image1.png')
plt.show()
```

```python
#Number of 15 - 25 year olds
len([i for i in age if i >= 15 and i <= 25])
print('Number of individuals between the ages of 15 and 25:',len(age))
```

```python
df = data.loc[(data['gender'] == 1) & (data['age'] == 30)]
print('Percentage of 30 year old males:',len(df)/len(data)*100,'%')
```

```python
df = data.loc[(data['age'] >= 25) & (data['age'] <= 54)]
print('Percentage of individuals between 25 and 54:',len(df)/len(data)*100)
```
